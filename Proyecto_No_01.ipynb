### **Problema 1:** 
#### Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.

```python
#Importamos la libreria pandas y numpy que se van a utilizar para los cuatro ejercicios
import pandas as pd
import numpy as np
```

```python
#Generamos las variables de día inicial y día final para introducir los últimos 5 días
dia_inicial = int(input('Introduce el dia inicial: '))
dia_final = int(input('Introduce el dia final: '))
#Definimos un diccionario vacio para el producto y mediante el ciclo for definimos el rango de días introducidos en las
#variables anteriores, el producto lo vamos a introducir como un dato de tipo string y a la vez podemos considerarlo un objeto, 
#por lo que no se utilizará el float como es el caso de las siguientes variables. Solicitamos a pd.Series que los productos 
#introducidos se muestren en serie. 
producto ={}
for i in range(dia_inicial, dia_final+1):
    producto[i] = input('Introduce el nombre del producto vendido en el dia ' + str(i) +': ')
    producto = pd.Series(producto, name='Producto')
#Generamos la variable ventas con un diccionario vacio que nos permita introducir la cantidad de productos vendidos del día 
#y mediante ciclo for generamos el rango de días definidos con los datos ingresados en el día inicial y final. Introducida
#la cantidad de producto vendido, solicitamos que los datos se generen en serie mediante el método pd.Series, etiquetamos
#las ventas como productos vendidos
ventas ={}
for i in range(dia_inicial, dia_final+1):
    ventas[i] = float(input('Introduce la cantidad de productos vendidos en el dia ' + str(i) +': '))
    ventas = pd.Series(ventas, name='Productos Vendidos')
#Generamos variable precio con un diccionario vacio que nos permita introducir el precio unitario para los cinco días definiendo
#que la variable precio sea de tipo float y nos arroje los datos en serie mediante pd.Series de pandas, etiquetando la variable
#precio como Precio Unitario. 
precio ={}
for i in range(dia_inicial, dia_final+1):
    precio[i] = float(input('Introduce el precio unitario ' + str(i) +': '))
    precio = pd.Series(precio, name='Precio Unitario')
#Generamos la variable precio descuento con ciclo for para obtener el precio con descuento del 5%.
    precio_descuento = (precio-(precio*0.05))
    precio_descuento = pd.Series(precio_descuento, name='Precio Descuento')
#De la misma forma generamos la variable total, la cual se define por la multiplicación de los productos vendidos y el precio 
#unitario para generar el total de ingresos, el cual pedimos que igualmente sea arrojado en serie mediante pd.Series. 
    total = (ventas*precio)
    total = pd.Series(total, name='Total Ingresos')
#Generamos la variable descuento, restando al total de ingresos los ingresos multiplicados por 0.05, y mediante el metodo 
#pd.Series solicitamos que arroje los datos en serie. 
    descuento = (total-(total*0.05))
    descuento = pd.Series(descuento, name='Ingresos con Descuento')
#Imprimimos los resultados en forma de serie como ya se habia especificado, y mediante el método pd.concat de tipo inner unimos las series:
#productos vendidos, precio unitario, Total Ingresos, e Ingresos con Descuento, con el fin de generar los datos en un
#DataFrame
print('Producto\n', producto)
print('Productos Vendidos\n', ventas)
print('Precio Unitario\n', precio)
print('Precio Descuento\n', precio_descuento)
print('Total Ingresos\n', total)
print('Ingresos con Descuento\n', descuento)
pd.concat([producto, ventas, precio, precio_descuento, total, descuento], axis = 1, join = "inner")
```

#### A continuación, generar nuevas columna:
#### 1. Columna 1 = Gasto/Ventas*100.
#### 2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.

#### 5. Obtener el total de gastos de febrero a abril

```python
#Primero es necesario generar el DataFrame con los datos proporcionados en el ejercicio
df = pd.DataFrame({'Mes':['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
                    'Ventas':[30500, 35600, 28300, 33900, 42500],
                    'Gastos':[22000, 23450, 18100, 35700, 32450]})
#Generamos la columna 1 y con el método apply obtenemos el porcentaje gastos sobre las ventas, realizando la operación en el eje1. 
df['Columna 1'] = df.apply(lambda x: round(x.Gastos/x.Ventas*100, 1), axis=1)
#Generamos la columna 2 y solicitamos que se imprima la etiqueta Meta Superada con la condicional if si el prefijo i, que corresponde
#a las ventas, es mayor a 30000. Por el contrario, si resulta menor a 30000 con else solicitamos que se imprima Meta no Superada
#y con el ciclo for se especifica que esta operación se realice en el rando de datos correspondiente a las ventas.
df['Columna 2'] = ['Meta Superada' if i>=30000 else 'Meta no Superada' for i in df['Ventas']]
#Otra forma es generando una lista vacia y con el ciclo for solicitamos que, si las ventas son mayores a 30000, se imprima Meta Superada
#en la celda que corresponda mediante el método append asociado a la lista, en caso contrario (else) se imprimirá Meta no Superada mediante
#el mismo método. Al finalizar, definimos la columna dos en el dataframe a partir de la lista metas. 
#metas = []
#for i in df['Ventas']:
#  if i > 30000: 
#    metas.append('Meta Superada')
#  else: 
#    metas.append('Meta no Superada')
#df['Columna 2'] = metas
#Para el caso de la columna 3 con el metodo apply solicitamos que se resten los gastos menos las ventas si los gastos son mayores a las ventas,
#caso contrario (else) solicitamos que la celda correspondiente a los valores donde los gastos no sean mayores a las ventas se imprima con '0.0'
# en el contexto del eje 1. 
df['Columna 3'] = df.apply(lambda x: x.Gastos - x.Ventas if x.Gastos > x.Ventas else '0.0', axis = 1)
#Para el caso de las ventas de los 3 últimos meses y los gastos de febrero a abril utilizamos la funcion .loc para acceder a valores
#específicos de filas o columnas en el dataframe. Mediante el método pd.Series de pandas, solicitamos que sume del rango 2 al 5 los valores 
#correspondientes a los ultimos 3 meses de la columna ventas. Asumismo, numeric_only=True lo utilizamos para que nos devuelva valores 
#numéricos de la columna de ventas, de lo contrario nos indicará una advertencia. Para el caso de los gastos de febrero a abril es el mismo
#proceso. El resultado serán dos filas: una para ventas y otra para gastos, pero separados uno de otro. 
#df.loc['Ventas ults 3 meses'] = pd.Series(df[2:5].sum(numeric_only=True), index=['Ventas'])
#df.loc['Gastos Feb-Abr'] = pd.Series(df[1:4].sum(numeric_only=True), index=['Gastos'])
#Opcionalmente, ambos resultados pueden agregarse en una fila: 'Total', mediante una sintaxis similar a la anterior, utilizando la función
#.loc y el método pd.Series. 
df.loc['Total'] = pd.Series([df.Ventas[2:5].sum(axis=0), df.Gastos[1:4].sum(axis=0)], index=['Ventas', 'Gastos'])
#Generamos una copia del dataframe para llenar los valores perdidos con 0.0. 
df_ventas=df.fillna(0.0)
df_ventas

```
